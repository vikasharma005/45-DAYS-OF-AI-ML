```python
def add_element(s):
    ele = input("Enter element to add: ")
    s.add(ele)
    print("Set after adding:", s)

def copy_set(s):
    backup = s.copy()
    print("Backup data:", backup)

def clear_set(s):
    s.clear()
    print("Empty set:", s)

def delete_element(s):
    l = list(s)
    print("Elements:", l)
    ind = int(input("Enter index value to remove element: "))
    if 0 <= ind < len(l):
        l.pop(ind)
        s = set(l)
        print("Set after deletion:", s)
    else:
        print("Invalid index.")

def union_set(s1, s2):
    union_result = s1.union(s2)
    print("Union of sets:", union_result)

def intersection_set(s1, s2):
    intersection_result = s1.intersection(s2)
    print("Intersection of sets:", intersection_result)

def difference_set(s1, s2):
    difference_result = s1.difference(s2)
    print("Difference of sets (s1 - s2):", difference_result)

def symmetric_difference_set(s1, s2):
    symmetric_difference_result = s1.symmetric_difference(s2)
    print("Symmetric Difference of sets:", symmetric_difference_result)

def difference_update_set(s1, s2):
    s1.difference_update(s2)
    print("Set after updating difference (s1 - s2):", s1)

def update_set(s1, s2):
    s1.update(s2)
    print("Set after updating with s2:", s1)

def is_subset(s1, s2):
    is_subset_result = s1.issubset(s2)
    print("Is s1 a subset of s2:", is_subset_result)

def is_superset(s1, s2):
    is_superset_result = s1.issuperset(s2)
    print("Is s1 a superset of s2:", is_superset_result)

if __name__ == "__main__":
    s1 = {'Allen', 'Finn', 'Luke', 'Mike', 'Sean'}
    s2 = {'Luke', 'Sean', 'Emily', 'Finn', 'Ella'}

    print("Options:")
    print("1. Add")
    print("2. Copy")
    print("3. Delete")
    print("4. Clear data")
    print("5. Union")
    print("6. Intersection")
    print("7. Difference")
    print("8. Symmetric Difference")
    print("9. Difference Update (s1 - s2)")
    print("10. Update (s1 + s2)")
    print("11. Is Subset")
    print("12. Is Superset")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        add_element(s1)
    elif choice == 2:
        copy_set(s1)
    elif choice == 3:
        delete_element(s1)
    elif choice == 4:
        clear_set(s1)
    elif choice == 5:
        union_set(s1, s2)
    elif choice == 6:
        intersection_set(s1, s2)
    elif choice == 7:
        difference_set(s1, s2)
    elif choice == 8:
        symmetric_difference_set(s1, s2)
    elif choice == 9:
        difference_update_set(s1, s2)
    elif choice == 10:
        update_set(s1, s2)
    elif choice == 11:
        is_subset(s1, s2)
    elif choice == 12:
        is_superset(s1, s2)
    else:
        print("Invalid option")
```

```python
def remove_duplicates_from_file(file_path):
    try:
        with open(file_path, "r") as file:
            elements = file.read().split()

        unique_elements = set(elements)

        with open(file_path, "w") as file:
            file.write(" ".join(unique_elements))

        print("Duplicates removed and unique elements written back to the file.")

    except FileNotFoundError:
        print("File not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    file_path = "/content/emp"
    remove_duplicates_from_file(file_path)
```
