**Multiple Condn Statements**

```python
a,b,c,d,e = 10,20,25,56,5
if a<b: ## 10<20  => True
  print(f"{b} Is Greater Than {a}") ## Execute
if b<c: ## 20<25 => True
  print(f"{c} Is Greater Than {b}") ## Execute
if c<d: ## 25<56 => True
  print(f"{d} Is Greater Than {c}") ## Execute
if d<e: ## 56<5 => False
  print(f"{e} Is Greater Than {d}")
else:
  print(f"{d} Is Greater Than {e}")  ## Execute
```

```python
a,b,c,d,e = 10,20,25,56,5
if a<b: ## 10<20  => True
  print(f"{b} Is Greater Than {a}") ## Execute
if b>c: ## 20>25 => False  ## Stop
  print(f"{b} Is Greater Than {c}")
if c<d: ## 25<56 => True
  print(f"{d} Is Greater Than {c}") ## Execute
if d<e: ## 56<5 => False
  print(f"{e} Is Greater Than {d}")
else:
  print(f"{d} Is Greater Than {e}")  ## Execute
```
