# Q1 Make a program about Marksheet

```python
def calculate_grade(percentage):
    if percentage >= 90:
        return "A+"
    elif percentage >= 80:
        return "A"
    elif percentage >= 70:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 50:
        return "D"
    else:
        return "F"

def create_marksheet():
    print("Enter marks for each subject (out of 100)")
    math = int(input("Math: "))
    science = int(input("Science: "))
    english = int(input("English: "))
    social_studies = int(input("Social Studies: "))
    computer_science = int(input("Computer Science:"))

    total_marks = math + science + english + social_studies + computer_science
    percentage = (total_marks / 500) * 100
    grade = calculate_grade(percentage)

    print("\n------ Marksheet ------")
    print(f"Math: {math}")
    print(f"Science: {science}")
    print(f"English: {english}")
    print(f"Social Studies: {social_studies}")
    print(f"Computer Science: {computer_science}")
    print("-----------------------")
    print(f"Total Marks: {total_marks}")
    print(f"Percentage: {percentage:.2f}%")
    print(f"Grade: {grade}")


print("***** Marksheet Generator *****")
create_marksheet()
```



This Python program generates a Marksheet for a student based on the marks obtained in five subjects: Math, Science, English, Social Studies, and Computer Science. The program comprises two functions: `calculate_grade(percentage)` and `create_marksheet()`.

### `calculate_grade(percentage)` Function:

This function takes the percentage obtained by the student as input and calculates the corresponding grade based on the grading system. The grading system used in this program is as follows:

- Percentage >= 90%: Grade is "A+"
- 80% <= Percentage < 90%: Grade is "A"
- 70% <= Percentage < 80%: Grade is "B"
- 60% <= Percentage < 70%: Grade is "C"
- 50% <= Percentage < 60%: Grade is "D"
- Percentage < 50%: Grade is "F"

### `create_marksheet()` Function:

This function is the core of the program and is responsible for gathering user input for the marks obtained in each subject and generating the Marksheet. Here's the step-by-step process:

1. The user is prompted to enter the marks obtained in each subject (out of 100). Input is taken using `int(input(...))` to ensure that the marks are treated as integers.

2. The total marks are calculated by adding the marks obtained in all five subjects.

3. The percentage obtained by the student is calculated by dividing the total marks by 500 (maximum marks for all subjects) and then multiplying by 100.

4. The `calculate_grade()` function is called with the percentage as an argument to determine the grade achieved by the student.

5. The program then displays the Marksheet, which includes the individual subject marks, total marks, percentage (formatted with two decimal places), and the calculated grade.



# Q2 Make a program about Bill Calculator

```python
def calculate_total_cost(items):
    total_cost = 0
    for item in items:
        total_cost += item["price"] * item["quantity"]
    return total_cost

def print_bill(items, total_cost):
    print("\n********** Bill **********")
    for item in items:
        print(f"{item['name']} x{item['quantity']}: ₹{item['price'] * item['quantity']}")
    print("--------------------------")
    print(f"Total Cost: ₹{total_cost}")
    print("***************************")

def bill_calculator():
    items = []
    while True:
        name = input("Enter item name (or 'done' to finish): ")
        if name.lower() == 'done':
            break
        price = float(input("Enter item price (₹): "))
        quantity = int(input("Enter item quantity: "))

        item = {"name": name, "price": price, "quantity": quantity}
        items.append(item)

    total_cost = calculate_total_cost(items)
    print_bill(items, total_cost)

print("***** Bill Calculator *****")
bill_calculator()
```



This Python program allows users to calculate a bill based on the quantity and price of items purchased. It comprises three functions: `calculate_total_cost(items)`, `print_bill(items, total_cost)`, and `bill_calculator()`.

### `calculate_total_cost(items)` Function:

This function takes a list of items as input and calculates the total cost of all the items purchased. It performs the following steps:

1. It initializes the `total_cost` variable to zero.
2. It iterates through each item in the provided `items` list.
3. For each item, it calculates the cost by multiplying the `price` of the item by its `quantity`.
4. The calculated cost is then added to the `total_cost`.
5. The function returns the final `total_cost`.

### `print_bill(items, total_cost)` Function:

This function takes two arguments: the list of items and the total cost. It displays the bill on the screen with item details and the total cost. Here's the process:

1. It prints a header section to indicate the start of the bill.
2. For each item in the `items` list, it prints the item name, quantity, and the calculated cost (price * quantity).
3. It prints a separator line to separate the item details from the total cost.
4. Finally, it prints the total cost of all items.

### `bill_calculator()` Function:

This function serves as the main function for the program and handles the user input for creating the bill. Here's the workflow:

1. The function initializes an empty list `items` to store the details of each item purchased.
2. It starts a while loop that keeps running until the user enters 'done' for the item name.
3. Inside the loop, the user is prompted to input the name, price, and quantity of each item.
4. A dictionary is created for each item with keys for `name`, `price`, and `quantity`, and the dictionary is appended to the `items` list.
5. When the user enters 'done' for the item name, the loop ends.
6. The function then calculates the total cost using the `calculate_total_cost()` function and calls the `print_bill()` function to display the bill.

```python
def print_menu():
    print("\n1. Length Conversion")
    print("2. Weight Conversion")
    print("3. Temperature Conversion")
    print("4. Volume Conversion")
    print("5. Time Conversion")
    print("6. Exit")

def length_conversion():
    print("1. Centimeters to Inches")
    print("2. Inches to Centimeters")
    print("3. Meters to Feet")
    print("4. Feet to Meters")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        cm = float(input("Enter length in centimeters: "))
        inches = cm * 0.393701
        print(f"{cm} centimeters is equal to {inches:.2f} inches.")
    elif choice == 2:
        inches = float(input("Enter length in inches: "))
        cm = inches / 0.393701
        print(f"{inches} inches is equal to {cm:.2f} centimeters.")
    elif choice == 3:
        meters = float(input("Enter length in meters: "))
        feet = meters * 3.28084
        print(f"{meters} meters is equal to {feet:.2f} feet.")
    elif choice == 4:
        feet = float(input("Enter length in feet: "))
        meters = feet / 3.28084
        print(f"{feet} feet is equal to {meters:.2f} meters.")
    else:
        print("Invalid choice.")

def weight_conversion():
    print("1. Kilograms to Pounds")
    print("2. Pounds to Kilograms")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        kg = float(input("Enter weight in kilograms: "))
        pounds = kg * 2.20462
        print(f"{kg} kilograms is equal to {pounds:.2f} pounds.")
    elif choice == 2:
        pounds = float(input("Enter weight in pounds: "))
        kg = pounds / 2.20462
        print(f"{pounds} pounds is equal to {kg:.2f} kilograms.")
    else:
        print("Invalid choice.")

def temperature_conversion():
    print("1. Celsius to Fahrenheit")
    print("2. Fahrenheit to Celsius")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        celsius = float(input("Enter temperature in Celsius: "))
        fahrenheit = (celsius * 9/5) + 32
        print(f"{celsius} Celsius is equal to {fahrenheit:.2f} Fahrenheit.")
    elif choice == 2:
        fahrenheit = float(input("Enter temperature in Fahrenheit: "))
        celsius = (fahrenheit - 32) * 5/9
        print(f"{fahrenheit} Fahrenheit is equal to {celsius:.2f} Celsius.")
    else:
        print("Invalid choice.")

def volume_conversion():
    print("1. Liters to Gallons")
    print("2. Gallons to Liters")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        liters = float(input("Enter volume in liters: "))
        gallons = liters * 0.264172
        print(f"{liters} liters is equal to {gallons:.2f} gallons.")
    elif choice == 2:
        gallons = float(input("Enter volume in gallons: "))
        liters = gallons / 0.264172
        print(f"{gallons} gallons is equal to {liters:.2f} liters.")
    else:
        print("Invalid choice.")

def time_conversion():
    print("1. Hours to Minutes")
    print("2. Minutes to Hours")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        hours = float(input("Enter time in hours: "))
        minutes = hours * 60
        print(f"{hours} hours is equal to {minutes:.2f} minutes.")
    elif choice == 2:
        minutes = float(input("Enter time in minutes: "))
        hours = minutes / 60
        print(f"{minutes} minutes is equal to {hours:.2f} hours.")
    else:
        print("Invalid choice.")

while True:
    print_menu()
    option = int(input("Enter your choice (1-6): "))

    if option == 1:
        length_conversion()
    elif option == 2:
        weight_conversion()
    elif option == 3:
        temperature_conversion()
    elif option == 4:
        volume_conversion()
    elif option == 5:
        time_conversion()
    elif option == 6:
        print("Exiting the unit converter.")
        break
    else:
        print("Invalid choice. Please try again.")
```

1. The program starts by defining a function `print_menu()` that displays the available conversion options.

3. Each conversion function takes user input for the value to be converted and performs the appropriate conversion based on the user's choice.
4. The main loop of the program runs indefinitely using a `while True` loop until the user chooses to exit (`6. Exit` option).
5. Inside the loop, the program prints the menu and asks the user for their choice (`option`).
6. Based on the user's choice, the program calls the corresponding conversion function.
7. After performing the conversion, the program returns to the menu, allowing the user to make another choice or exit the program.

# Q4 Make program about BMI calculator
            
   ##### BMI formula: BMI = weight (kg) / (height (m) * height (m))


```python
def calculate_bmi(weight, height):
    # BMI formula: BMI = weight (kg) / (height (m) * height (m))
    bmi = weight / (height * height)
    return bmi

def get_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obese"

def main():
    print("BMI Calculator")
    print("----------------")

    try:
        weight = float(input("Enter your weight (kg): "))
        height = float(input("Enter your height (m): "))

        bmi = calculate_bmi(weight, height)
        category = get_category(bmi)

        print(f"\nYour BMI is: {bmi:.2f}")
        print(f"You are categorized as: {category}")

    except ValueError:
        print("Invalid input. Please enter numeric values for weight and height.")

if __name__ == "__main__":
    main()
```



The BMI (Body Mass Index) Calculator is designed to calculate the BMI of a person based on their weight and height. It then categorizes the BMI into different weight categories, such as "Underweight," "Normal weight," "Overweight," or "Obese."

## Program Logic
1. The program defines two functions: `calculate_bmi(weight, height)` and `get_category(bmi)`.
    - `calculate_bmi(weight, height)`: This function takes the weight (in kilograms) and height (in meters) as input and calculates the BMI using the formula `BMI = weight / (height * height)`. It returns the calculated BMI value.
    - `get_category(bmi)`: This function takes the BMI value as input and determines the weight category based on predefined ranges. It returns the corresponding weight category.

2. The `main()` function is defined as the entry point of the program.
    - It displays the program title and prompts the user to enter their weight and height.
    - It converts the user input into floating-point numbers using `float()` function.
    - It calls the `calculate_bmi()` function with the user-provided weight and height to calculate the BMI.
    - It calls the `get_category()` function with the calculated BMI to determine the weight category.

3. The program uses a `try-except` block to handle potential ValueError exceptions if the user enters non-numeric input for weight or height.

4. Finally, the program displays the calculated BMI value and the corresponding weight category.

```python
def generate_formal_letter(sender_name, sender_address, recipient_name, date):
    formal_letter = f"""
    {sender_name}
    {sender_address}
    {date}

    {recipient_name}
    Address Line 1
    Address Line 2

    Dear {recipient_name},

    I hope this letter finds you in good health and high spirits.

    I am writing to address an important matter regarding [purpose of the letter].

    [Content of the formal letter]

    Thank you for your attention to this matter. I look forward to hearing from you soon.

    Sincerely,
    {sender_name}
    """

    return formal_letter

def generate_informal_letter(sender_name, recipient_name, date):
    informal_letter = f"""
    Dear {recipient_name},

    Hey there! I hope you're doing well.

    Just wanted to drop you a line to talk about [subject of the letter].

    [Content of the informal letter]

    Well, that's all for now. Let's catch up soon!

    Best regards,
    {sender_name}
    """

    return informal_letter

def main():
    print("Letter Generator")
    print("-----------------")

    sender_name = input("Your Name: ")
    sender_address = input("Your Address: ")
    recipient_name = input("Recipient's Name: ")
    date = input("Date (e.g., July 20, 2023): ")

    print("\nFormal Letter:")
    formal_letter = generate_formal_letter(sender_name, sender_address, recipient_name, date)
    print(formal_letter)

    print("\nInformal Letter:")
    informal_letter = generate_informal_letter(sender_name, recipient_name, date)
    print(informal_letter)

if __name__ == "__main__":
    main()
```



This Python program generates both formal and informal letter templates based on user input.

#### Main Function

The `main()` function serves as the entry point of the program. It prompts the user to input the required information for both formal and informal letters, such as the sender's name, address, recipient's name, and date.

1. The user is asked to enter their name, address, recipient's name, and the date of the letter.

2. After receiving the input, the program calls the `generate_formal_letter()` function with the provided information and stores the generated formal letter template in the `formal_letter` variable.

# Q6 Make a Function to create 5 lists using while loop

```python
def create_five_lists():
    lists = []
    count = 1

    while count <= 5:
        new_list = []
        size = int(input(f"Enter the size of list {count}: "))
        for i in range(size):
            element = input(f"Enter element {i+1} of list {count}: ")
            new_list.append(element)
        lists.append(new_list)
        count += 1

    return lists

lists = create_five_lists()
print("Lists created:")
for i, lst in enumerate(lists, 1):
    print(f"List {i}: {lst}")
```




This Python function, `create_five_lists()`, is designed to create five lists with user-defined elements using a while loop. Here's how the function works:

1. The function initializes an empty list called `lists` to store the five lists created by the user. It also sets a counter variable `count` to 1 to keep track of the current list being created.

2. The function enters a while loop that continues until `count` reaches 5. This ensures that the function creates exactly five lists as required.

3. Inside the loop, the function prompts the user to input the size (number of elements) for the current list being created using `int(input("Enter the size of list {count}: "))`. The `int()` function is used to convert the user's input into an integer.

4. For each list, the function initializes an empty list called `new_list` to store the elements provided by the user.

5. It then runs a for loop using `for i in range(size)` to iterate `size` number of times. In each iteration, the function asks the user to input an element for the current list using `input("Enter element {i+1} of list {count}: ")`, where `i+1` is used to display the element number starting from 1 instead of 0.

6. The user-provided element is then appended to the `new_list` using the `append()` method.

7. Once all elements are added to the `new_list`, the function appends the entire `new_list` to the main `lists` list, effectively adding the current list to the list of lists.

8. The `count` is then incremented by 1 to move on to the next list.

9. The loop continues until `count` becomes 5, creating all five lists.

10. Finally, the function returns the list of lists (`lists`) containing the five lists created by the user.

# Q8 Make a login page and get data after 5 steps:-
        1. LogIn
        2. OTP
        3. Captcha
        4. Security Question
        5. Key

```python
import random
import getpass
import colorama
from colorama import Fore, Style

# Initialize colorama to work on Windows
colorama.init(autoreset=True)

# Set autoreset to False to avoid escape codes in input prompts
colorama.init(autoreset=True, convert=False)

# Predefined username and password (you can change these)
USERNAME = "vikas005"
PASSWORD = "pass123"

def login():
    print(f"{Fore.CYAN}{Style.BRIGHT}--- Login Page --- {Style.RESET_ALL}")
    username = input("Enter your username: ")
    password = getpass.getpass("Enter your password: ")

    # Check credentials (you'd compare with stored data in a real application)
    if username == USERNAME and password == PASSWORD:
        print(f"{Fore.GREEN} Login successful! Welcome, {username}.")
        return True
    else:
        print(f"{Fore.RED} Invalid credentials. Please try again.")
        return False

def generate_otp():
    otp = str(random.randint(1000, 9999))
    print(f"{Fore.YELLOW}Generated OTP: {otp}")
    return otp

def verify_otp():
    print(f"{Fore.CYAN}{Style.BRIGHT} --- OTP Verification --- {Style.RESET_ALL}")
    otp = generate_otp()
    user_otp = input(f"{Style.RESET_ALL}{Fore.YELLOW} Enter the OTP sent to your registered mobile number: ")

    if otp == user_otp:
        print(f"{Fore.GREEN} OTP verification successful.")
        return True
    else:
        print(f"{Fore.RED} Invalid OTP. Please try again.")
        return False

def generate_captcha():
    captcha_length = 6
    captcha_characters = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789"
    captcha = "".join(random.choice(captcha_characters) for _ in range(captcha_length))
    return captcha

def verify_captcha():
    print(f"{Fore.CYAN}{Style.BRIGHT} --- Captcha Verification --- {Style.RESET_ALL}")
    captcha_code = generate_captcha()
    print(f"{Style.RESET_ALL}{Fore.MAGENTA} Generated Captcha: {captcha_code}")
    user_captcha = input(f"{Style.RESET_ALL}{Fore.MAGENTA} Enter the Captcha code (case-sensitive): ")

    if user_captcha == captcha_code:
        print(f"{Fore.GREEN}Captcha verification successful.")
        return True
    else:
        print(f"{Fore.RED}Invalid Captcha. Please try again.")
        return False

def verify_security_question():
    print(f"{Fore.CYAN}{Style.BRIGHT}--- Security Question --- {Style.RESET_ALL}")
    security_question = "What is your pet's name?"
    correct_answer = "tom"
    user_answer = input(f"{Fore.YELLOW} {security_question}: ")

    if user_answer == correct_answer:
        print(f"{Fore.GREEN}Security question verification successful. ")
        return True
    else:
        print(f"{Fore.RED}Incorrect answer. Please try again. ")
        return False

def verify_key():
    print(f"{Fore.CYAN}{Style.BRIGHT} --- Key Verification --- {Style.RESET_ALL}")
    correct_key = "1234"
    user_key = input(f"{Style.RESET_ALL} {Fore.BLUE} Enter the secret key:   ")

    if user_key == correct_key:
        print(f"{Fore.GREEN} Key verification successful.")
        return True
    else:
        print(f"{Fore.RED} Invalid key. Please try again. ")
        return False

if __name__ == "__main__":
    login_successful = login()
    if not login_successful:
        exit()

    otp_verified = verify_otp()
    if not otp_verified:
        exit()

    captcha_verified = verify_captcha()
    if not captcha_verified:
        exit()

    security_question_verified = verify_security_question()
    if not security_question_verified:
        exit()

    key_verified = verify_key()
    if not key_verified:
        exit()

    # If all steps are successful, proceed to the next part of the application.
    print(f"{Fore.GREEN} Login successful! You are now logged in.")
```



This is a simple Python script that simulates a basic login process with multiple steps, including OTP verification, Captcha verification, Security Question, and Key verification. The `colorama` library is used to add some color to the terminal output.

#### Program Flow:

1. Import necessary libraries: The program starts by importing the required libraries, including `random` for generating OTP and Captcha, `getpass` for secure password input, and `colorama` for colorful terminal text.

2. Define predefined username and password: The program sets a predefined `USERNAME` and `PASSWORD`. In a real application, these credentials would be stored securely, e.g., in a database.

3. Implement the login function: The `login()` function is defined to handle the login process. It displays a login prompt and asks the user to input their username and password.

4. Secure password input: The `getpass.getpass()` function is used to safely get the password input from the user without displaying it on the screen.

5. Validate credentials: The user's input is then compared with the predefined username and password. If they match, a success message is displayed; otherwise, an error message is shown.

6. Implement the OTP verification function: The `generate_otp()` function generates a random 4-digit OTP, and the `verify_otp()` function asks the user to enter the OTP sent to their registered mobile number for verification.

7. Implement the Captcha verification function: The `generate_captcha()` function generates a random 6-character alphanumeric Captcha, and the `verify_captcha()` function asks the user to enter the Captcha code displayed on the screen for verification.

8. Implement the Security Question function: The `verify_security_question()` function asks the user a security question and verifies their answer against a predefined correct answer.

9. Implement the Key verification function: The `verify_key()` function asks the user to enter a secret key and verifies it against a predefined correct key.

10. Run the program: The main part of the program runs the login process and each verification step. If all steps are successful, a final success message is displayed, indicating that the user is now logged in.

```python
from tabulate import tabulate
import colorama
from colorama import Fore, Style

# Initialize colorama to work on Windows
colorama.init(autoreset=True)

# Sample data for the table
data = [
    ["Name", "Age", "Country"],
    ["John", 30, "USA"],
    ["Alice", 25, "Canada"],
    ["Bob", 22, "UK"],
]

# Table headers and data, aligned center and colored
table = tabulate(data, headers="firstrow", tablefmt="grid")
colored_table = {Fore.MAGENTA}{Style.BRIGHT}f"{table}{Style.RESET_ALL}"

print(colored_table)
```

# Q10 Make a program about the Income Tax Slab

```python
def calculate_income_tax(income):
    if income <= 250000:
        tax = 0
    elif income <= 500000:
        tax = (income - 250000) * 0.05
    elif income <= 1000000:
        tax = 250000 * 0.05 + (income - 500000) * 0.2
    else:
        tax = 250000 * 0.05 + 500000 * 0.2 + (income - 1000000) * 0.3
    return tax

print("Income Tax Calculator (Assessment Year 2023-2024)")
print("------------------------------------------------")
income = float(input("Enter your annual income: "))

if income < 0:
    print("Invalid input. Income cannot be negative.")
else:
    tax = calculate_income_tax(income)
    print(f"Your income tax for the year is: ₹ {tax:.2f}")
```



This is a simple Python program that calculates income tax based on the income tax slab rates for the assessment year 2023-2024 in India.
The user is prompted to enter their annual income, and the program computes the income tax accordingly.

## How the Program Works

1. The program defines a function `calculate_income_tax(income)` that takes the income as input and calculates the income tax based on the given tax slab rates.
2. The tax slab rates for the assessment year 2023-2024 are as follows:
   - Income up to ₹250,000: Tax rate is 0%
   - Income above ₹250,000 up to ₹500,000: Tax rate is 5%
   - Income above ₹500,000 up to ₹1,000,000: Tax rate is 20%
   - Income above ₹1,000,000: Tax rate is 30%
3. The program then takes user input for their annual income using `float(input("Enter your annual income: "))`.
4. If the user enters a negative income value, the program displays an error message indicating that the income cannot be negative.
5. Otherwise, the program calls the `calculate_income_tax()` function with the provided income and calculates the income tax based on the tax slab rates.
6. The program then prints the calculated income tax with the message "Your income tax for the year is: ₹ {tax:.2f}".




# Q11 How can we recall or reassign a variable after deleting it, especially when the variable name is denoted by the underscore (`_`)?

### Underscore Variable (_) / Hidden Variable

In Python, the underscore (`_`) is a special variable used as a placeholder for the result of the last expression in an interactive interpreter session. It's often referred to as a **"hidden variable"** because it's automatically set by the interpreter and not intended to be used explicitly by the programmer.


1. `>>> _ = 12`: In an interactive interpreter session, we set the `_` variable to the value 12.

2. `>>> del(_):` Next, we delete the `_` variable using `del(_)`. This removes the reference to the value 12, and the `_` variable no longer exists.

3. `>>> _ = 30`: Attempting to reassign a value to `_`, we try `_ = 30`.



```python
# Assign a value to the variable _
_ = 12

_
```

```python
# Delete the variable _
del _
```

The `_` variable in Python serves as a special variable for capturing the result of the last expression in an interactive interpreter session. Once we delete the `_` variable, we cannot recall or reassign it automatically. It's called a **"hidden variable"** because its behavior is managed by the interpreter, and ***it's not intended for explicit use by the programmer.***

If we want to **recall or reassign the value** from the previous expression, you need to save it in a different variable before deleting the original one. Here's an example of how you can achieve this:


```python
_ = 12

# Save the value in a different variable
saved_value = _

# Delete the original variable _
del _

# Now, we can reassign the saved value to a new variable named _
_ = 30

# Now, we can use the variable _
print(_)
```

# Q12 When we set `a1` as `10j`, it gives the output `10j`, representing a    complex number with a real part of 0 and an imaginary part of 10. However, when we set `a1` as `-10j`, it gives the output `-0-10j` with negative digits for the imaginary part. Why does it display negative digits in this case?

### **Complex Numbers in Python**

In Python, we can work with complex numbers by using the imaginary unit `j`. When you set a variable like `a1 = 10j`, we are creating a complex number with a real part of 0 and an imaginary part of 10.

- `a1 = 10j`: This represents a complex number with `0` as the real part and `10` as the imaginary part.

Now, when we set `a1 = -10j`, we are creating another complex number, but this time with a negative imaginary part. Python displays this negative imaginary part as `-0-10j`.

- `a1 = -10j`: This represents a complex number with `0` as the real part and `-10` as the imaginary part.

The appearance of `-0` instead of just `0` is a visual representation used to distinguish between positive and negative zero. However, mathematically, `-0` is equivalent to `0`. So, whether it's `-0-10j` or `0-10j`, the value of the complex number is the same, with a real part of `0` and an imaginary part of `-10`.

In most cases, the presence of `-0` doesn't affect the actual value or calculation of the complex number; it's just a way Python chooses to display the number.
