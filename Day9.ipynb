The error `IndexError: list index out of range` occurs because we are modifying the list `stud` while iterating over it using a `for` loop with the `remove` method. When we remove an element from the list, its size decreases, and the index of the elements after the removed element is updated accordingly. This causes an index out of range error when the loop reaches the end of the list after some elements have been removed.

Let's walk through what happens in each iteration of the loop:

1. Initial list: `['Mike', 'Luke', 'Jack', 'Peter', 'Finn', 'Nile', 'Allen', 'Jacob']`
2. `a = 0`: The element at index 0 is `'Mike'`, which stays in the list.
3. `a = 1`: The element at index 1 is `'Luke'`, and since `1 % 2` is not equal to 0, `'Luke'` is removed from the list. The list becomes `['Mike', 'Jack', 'Peter', 'Finn', 'Nile', 'Allen', 'Jacob']`.
4. `a = 2`: The element at index 2 is now `'Peter'`, which was previously at index 3. Since `'Luke'` was removed in the previous iteration, the elements are shifted, and `'Peter'` takes its place at index 2.
5. `a = 3`: The element at index 3 is now `'Finn'`, which was previously at index 4. Again, due to the shifting, `'Finn'` takes its place at index 3.
6. `a = 4`: The element at index 4 is now `'Nile'`, which was previously at index 5. The elements keep shifting, causing `'Nile'` to take its place at index 4.

At this point, the loop has already gone beyond the length of the updated list, and it tries to access an element at an index that doesn't exist, resulting in the `IndexError`.

To avoid this error, we should not modify the list while iterating over it. Instead, we can create a new list containing the elements we want to keep. This way, the original list remains unchanged, and we get the desired result without encountering an `IndexError`.

```python
def even_numbers_while(limit):
    total = 0
    current_number = 2
    while current_number <= limit:
        total += current_number
        current_number += 2
    return total

def even_numbers_for(limit):
    total = 0
    for number in range(2, limit + 1, 2):
        total += number
    return total

limit = int(input("Enter the upper limit: "))


result_while = even_numbers_while(limit)
print("Sum of even numbers using while loop:", result_while)


result_for = even_numbers_for(limit)
print("Sum of even numbers using for loop:", result_for)
```
