# Q2 Find Special Char , Number , Char ,Bool And Emoji From Given Data.

    Data = "Sean@gmail.comMike Is ðŸ˜€.Password Is 25166@##.Okay!!100True"

```python
data = "Sean@gmail.comMike Is ðŸ˜€.Password Is 25166@##.Okay!!100True"

special_char_count = 0
number_count = 0
char_count = 0
bool_count = 0
emoji_count = 0


def is_emoji(char):
    emoji_range = (0x1F600, 0x1F64F)
    return ord(char) in range(*emoji_range)


for char in data:
    if char.isalnum():
        if char.isalpha():
            char_count += 1
        else:
            number_count += 1
    elif char in "!@#$%^&*()_-+=<>?/,.:;{}[]|\\":
        special_char_count += 1
    elif char in ("True", "False"):
        bool_count += 1
    elif is_emoji(char):
        emoji_count += 1

print("Special Characters:", special_char_count)
print("Numbers:", number_count)
print("Characters:", char_count)
print("Booleans:", bool_count)
print("Emojis:", emoji_count)
```

# Q3 Encode A Msg Using Input Function With Options :

```python
def encode_message(msg, key, password):
    if password == "pass123":  # Replace "your_password_here" with the desired password
        encoded_msg = ""
        for char in msg:
            encoded_char = chr(ord(char) + key)
            encoded_msg += encoded_char
        return encoded_msg
    else:
        return "Password incorrect. Message encoding failed."

def decode_message(encoded_msg, key, password):
    if password == "pass123":
        decoded_msg = ""
        for char in encoded_msg:
            decoded_char = chr(ord(char) - key)
            decoded_msg += decoded_char
        return decoded_msg
    else:
        return "Password incorrect. Message decoding failed."

def main():
    while True:
        print("\nOptions:")
        print("1. Encode a message")
        print("2. Decode a message")
        print("3. Exit")

        choice = input("\nEnter your choice (1, 2, or 3): ")

        if choice == "1":
            msg = input("\n\nEnter the message to encode: ")
            key = int(input("Enter the key (an integer): "))
            password = input("Enter the password: ")
            encoded_msg = encode_message(msg, key, password)
            print("Encoded message:", encoded_msg)
        elif choice == "2":
            encoded_msg = input("\n\nEnter the message to decode: ")
            key = int(input("Enter the key (an integer): "))
            password = input("Enter the password: ")
            decoded_msg = decode_message(encoded_msg, key, password)
            print("Decoded message:", decoded_msg)
        elif choice == "3":
            print("\n\nExiting the program.")
            break
        else:
            print("\\nInvalid choice. Please select a valid option (1, 2, or 3).")

if __name__ == "__main__":
    main()
```

# Q4 Make A Program Using While And For Loop.

```python
def even_numbers_while(limit):
    total = 0
    current_number = 2
    while current_number <= limit:
        total += current_number
        current_number += 2
    return total

def even_numbers_for(limit):
    total = 0
    for number in range(2, limit + 1, 2):
        total += number
    return total

limit = int(input("Enter the upper limit: "))


result_while = even_numbers_while(limit)
print("Sum of even numbers using while loop:", result_while)


result_for = even_numbers_for(limit)
print("Sum of even numbers using for loop:", result_for)
```

# **Explaination :**

This Python code demonstrates two functions, `even_numbers_while` and `even_numbers_for`, that calculate the sum of even numbers up to a given upper limit. The code then prompts the user to enter the upper limit and calculates the sum using both functions.


1. `even_numbers_while(limit)`: This function calculates the sum of even numbers using a `while` loop. It takes the `limit` as an input, and `total` is initialized to 0. The `current_number` is initialized to 2 because it is the first even number. The function then iterates over the even numbers using the `while` loop until `current_number` exceeds the `limit`. It adds each even number to the `total` and increments `current_number` by 2 in each iteration. Finally, it returns the `total` sum.

2. `even_numbers_for(limit)`: This function calculates the sum of even numbers using a `for` loop. It also takes the `limit` as an input, and `total` is initialized to 0. The function uses the `range` function with a step of 2 to generate a sequence of even numbers from 2 up to `limit` (inclusive). It iterates over this sequence using the `for` loop, adding each even number to the `total`. Finally, it returns the `total` sum.

3. The user is prompted to enter the `limit` (the upper bound) for calculating the sum of even numbers.

4. The code then calls both functions, passing the user-provided `limit`, and stores the results in `result_while` and `result_for`, respectively.

5. The sums calculated using the `while` loop and `for` loop are printed to the console.



# Q5 Make A Program To Find Any Word From HTML And Text File.

```python
import re

def main():
    file_path = input("Enter the file path: ")
    word_to_find = input("Enter the word to find: ")

    try:
        with open(file_path, 'r') as file:
            content = file.read()

        # Regular expression to find the word in the content
        matches = re.findall(r'\b' + re.escape(word_to_find) + r'\b', content, re.IGNORECASE)

        if matches:
            print(f"The word '{word_to_find}' is found in the file {len(matches)} times:")
            for idx, match in enumerate(matches, 1):
                print(f"{idx}. {match}")
        else:
            print(f"The word '{word_to_find}' is not found in the file.")
    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
    except Exception as e:
        print("An error occurred:", e)

if __name__ == "__main__":
    main()
```

2. The regular expression for finding the word is now constructed directly within the `re.findall` function call.

3. The output messages are adjusted to be more concise.

4. The variable names and overall structure of the program remain the same.



# Q6 str1 = "abnjeropsuilvsdabnjer"
    Find The Unique Values From Given String.

    Append The Single Char In A List.

    Make Two Words.

    Find The Vowels From Given String.
    
    Find The Consonant From Given String.

```python
str1 = "abnjeropsuilvsdabnjer"

# Find unique values from the string
unique_values = [char for char in str1 if str1.count(char) == 1]
print("Unique values:", unique_values)

# Make two words from the string
mid = len(str1) // 2
word1 = str1[:mid]
word2 = str1[mid:]
print("Word 1:", word1)
print("Word 2:", word2)

# Find the vowels from the string
vowels = set('aeiouAEIOU')
vowels_list = [char for char in str1 if char in vowels]
print("Vowels:", vowels_list)

# Find the consonants from the string
consonants_list = [char for char in str1 if char.isalpha() and char not in vowels]
print("Consonants:", consonants_list)
```

# **Explaination :**

1. `unique_values`: This list comprehension iterates over each character in the string `str1` and checks if the count of that character in the string is equal to 1. If so, it means the character is unique, and it is added to the `unique_values` list.

2. `word1` and `word2`: These variables split the string `str1` into two words by finding the midpoint (using integer division `//`) and assigning the first half to `word1` and the second half to `word2`.

3. `vowels`: This set contains all the vowels, both lowercase and uppercase.

4. `vowels_list`: This list comprehension iterates over each character in `str1` and checks if the character is present in the `vowels` set. If so, it adds the character to the `vowels_list`.

5. `consonants_list`: This list comprehension iterates over each character in `str1` and checks if the character is an alphabetic character (using `char.isalpha()`) and is not a vowel (i.e., not present in the `vowels` set). If both conditions are met, it adds the character to the `consonants_list`.


# Q7 Marks = {"Sean":560,"Mike":520,"Jenny":595,"Luke":500,"Finn":555}
    Take A Input And Get The Marks of Stud(Key) if Exist.
    Get The Marks With Percentage , Here Total Marks Is 600.
    Find The Max And Min Percentage.
    Find The Mean And Median.

```python
def get_per(marks, total_marks=600):
    return (marks / total_marks) * 100

def max_per(marks_dict):
    return max(marks_dict.values())

def min_per(marks_dict):
    return min(marks_dict.values())

def mean_per(marks_dict):
    total_percentage = sum(marks_dict.values())
    return total_percentage / len(marks_dict)

def median_per(marks_dict):
    sorted_marks = sorted(marks_dict.values())
    n = len(sorted_marks)
    if n % 2 == 0:
        median = (sorted_marks[n//2 - 1] + sorted_marks[n//2]) / 2
    else:
        median = sorted_marks[n//2]
    return median

Marks = {"Sean": 560, "Mike": 520, "Jenny": 595, "Luke": 500, "Finn": 555}


student_name = input("Enter the student's name to get marks: ")
if student_name in Marks:
    student_marks = Marks[student_name]
    print(f"{student_name}'s marks: {student_marks}")
    print(f"Percentage: {get_per(student_marks):.2f}%")
else:
    print(f"Student '{student_name}' not found in the records.")

# Find the max and min percentage
max_percentage = max_per(Marks)
min_percentage = min_per(Marks)
print(f"Max Percentage: {max_percentage:.2f}%")
print(f"Min Percentage: {min_percentage:.2f}%")

# Find the mean and median percentage
mean_percentage = mean_per(Marks)
median_percentage = median_per(Marks)
print(f"Mean Percentage: {mean_percentage:.2f}%")
print(f"Median Percentage: {median_percentage:.2f}%")
```

# **Expalination : **

1. Function Definitions:
   - `get_per(marks, total_marks=600)`: This function calculates the percentage obtained by dividing the marks by the total marks (defaulted to 600) and multiplying by 100. It returns the percentage as a floating-point value.

   - `max_per(marks_dict)`: This function takes a dictionary `marks_dict` as input, which contains student names as keys and their corresponding marks as values. It finds the maximum marks from the values in the dictionary and returns the maximum percentage.

   - `min_per(marks_dict)`: Similar to `max_per`, this function finds the minimum marks from the values in the dictionary and returns the minimum percentage.

   - `mean_per(marks_dict)`: This function calculates the mean percentage by adding all the percentages in the dictionary and dividing by the number of students (the length of the dictionary).

   - `median_per(marks_dict)`: This function calculates the median percentage from the values in the dictionary. It first sorts the percentages and then determines the median based on whether the number of students is even or odd.

2. Marks Dictionary:
   - The `Marks` dictionary contains the names of students as keys and their corresponding marks as values. This data represents the marks obtained by five students.

3. Input and Display:
   - The code prompts the user to input the name of a student to get their marks.
   - If the provided student name exists in the `Marks` dictionary, the corresponding marks and percentage are displayed using the `get_per` function.
   - If the provided student name is not found in the dictionary, it informs the user that the student is not in the records.

4. Finding Maximum and Minimum Percentage:
   - The `max_per` and `min_per` functions are used to find the maximum and minimum percentages, respectively, from the `Marks` dictionary.

5. Finding Mean and Median Percentage:
   - The `mean_per` function calculates the mean percentage of all the students' marks in the `Marks` dictionary.
   - The `median_per` function calculates the median percentage of all the students' marks. It sorts the percentages, checks if the number of students is even or odd, and computes the median accordingly.
